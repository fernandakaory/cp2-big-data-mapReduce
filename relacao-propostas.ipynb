## Celso Russomano - Fernanda

```python
import re

# Ler stopwords do arquivo
with open("stopwords.txt", "r", encoding="utf-8") as f:
    stopwords = set(f.read().split())

# Ler arquivo TXT principal
with open("propostas/CELSO-RUSSOMANNO.txt", "r", encoding="utf-8") as file:
    texto = file.read()

# Limpeza
texto = texto.lower()
texto = re.sub(r'[^\w\s]', '', texto)

palavras = texto.split()

# Remover stopwords
palavras_filtradas = [p for p in palavras if p not in stopwords]

print(palavras_filtradas[:20])
```

```python
from collections import defaultdict

# MAP
mapped = [(palavra, 1) for palavra in palavras_filtradas]

# REDUCE
frequencia = defaultdict(int)

for palavra, count in mapped:
    frequencia[palavra] += count

# Ordenar por frequência
frequencia_ordenada = sorted(frequencia.items(), key=lambda x: x[1], reverse=True)

# Mostrar top 10
top_15 = frequencia_ordenada[:15]
print(top_15)
```

```python
with open("resultados/resultado_frequencias1.txt", "w", encoding="utf-8") as f:
    for palavra, freq in frequencia_ordenada:
        f.write(f"{palavra}: {freq}\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(top_15, columns=["Palavra", "Frequência"])

plt.figure()
plt.bar(df["Palavra"], df["Frequência"], color="#db72bf")

plt.title("Palavras mais frequentes no plano de governo - CELSO RUSSOMANNO")
plt.xlabel("Palavras")
plt.ylabel("Frequência")

plt.xticks(rotation=90)

plt.show()
```

A partir do gráfico de barras, é possível observar que a palavra mais frequente no plano de governo de Celso Russomanno é “ampliar”, o que indica uma forte intenção de expandir programas, serviços e benefícios voltados à população paulistana. Esse foco em crescimento também aparece em outras palavras com alta frequência, como “criar”, “criação” e “implantar”, que mostram que o plano não se limita a melhorar o que já existe, mas também propõe novas iniciativas.

Além disso, a presença da palavra “saúde” entre as mais frequentes indica que essa é uma das principais áreas de prioridade. Isso sugere que o candidato pretende investir na ampliação e melhoria dos serviços de saúde, possivelmente com a criação de novas unidades e expansão do atendimento.

Outras palavras como “rede”, “centro” e “áreas” indicam uma preocupação com a estrutura e organização dos serviços públicos. Já termos como “administração” e “municipal” mostram que há um foco na gestão pública local, ou seja, na forma como essas ações serão organizadas e executadas dentro do município.

A palavra “população” reforça o foco social do plano, indicando que as propostas são voltadas diretamente para atender as necessidades das pessoas e melhorar a qualidade de vida.

Por outro lado, aparecem também os termos “1” e “2”, que, apesar de frequentes, não têm relevância para a análise, pois estão relacionados à organização do texto (como tópicos ou listas), e não ao conteúdo das propostas.

De forma geral, a análise mostra que o plano de governo está voltado para a expansão de serviços, criação de novas políticas públicas e melhoria da gestão municipal, com destaque para a área da saúde e para o atendimento da população.
